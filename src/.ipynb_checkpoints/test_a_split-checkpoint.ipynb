```python
import pandas as pd
import numpy as np
import os, sys, re

reads_28_place = "/cluster/bh0085/shortreads/190328Gif"
reads_29_place = "/cluster/bh0085/shortreads/190329Gif"
sequencing_info_file = "../data/smallmol2 sequencing directory contents - reads_28.csv"
```

```python
REDUCED_LIB_DESIGN = pd.read_csv("~/data/reduced_lib_design.csv")
REDUCED_LIB_DESIGN["Name"] = REDUCED_LIB_DESIGN["Identifier number"].apply(lambda x: int(x)) - 1
REDUCED_LIB_DESIGN.set_index("Name")
```

```python
info = pd.read_csv(sequencing_info_file,names=["Code"])

reads_28_files = os.listdir(reads_28_place)
reads_29_files = os.listdir(reads_29_place)

info = pd.concat([info,info.apply(lambda x:pd.Series( re.compile("(?P<Name>D[^\s]+) : (?P<Description>[^()]*)\((?P<Index>[^)]*)\)").search(x["Code"]).groupdict()),axis=1)],axis=1)
locations = pd.concat([info.Name.apply(lambda x: [e for e in reads_29_files if x in e]).apply(lambda x:os.path.join(reads_29_place,x[0]) if len(x)>0 else None).rename("dir29"),
           info.Name.apply(lambda x: [e for e in reads_28_files if x in e]).apply(lambda x:os.path.join(reads_28_place,x[0]) if len(x)>0 else None).rename("dir28")],axis=1)

locations2 = locations.apply(lambda x: x.dir29 if x.dir29 else x.dir28,axis = 1)
info["reads_location"] = locations2
info["fastq"] = info.apply(lambda x:[fn for fn in os.listdir(x.reads_location) if fn[-5:]=="fastq"][0],axis=1)
info["fastq_path"] = info["reads_location"] +"/"+ info["fastq"]
```

```python
def rc(s):
    return "".join([{"N":"N","A":"T","G":"C","T":"A","C":"G"}[l] for l in s][::-1])
```

```python
seqs = []
for fpath in info["fastq_path"].iloc[3:]:
    
    num_bad_q = 0
    num_total = 0
    
    with open(fpath) as f:
        for i,l in enumerate(f):
            if i%4 == 0:
                h1 = l.strip()
                continue
            if i%4 == 1:
                r1 = l.strip()
                continue
            if i%4 == 3:
                num_total+=1
                qs = l.strip()
                quals = [ord(s)-33 for s in qs]
                if np.mean(quals) < 32:        
                    num_bad_q += 1
                    continue
                #print( r1)
                #print(np.mean(quals))
                #print(REDUCED_LIB_DESIGN["Designed gRNA (NGG orientation, 19 or 20)"].apply(lambda x : x[:10] in rc(r1)).nonzero())
                seqs.append(r1)
                if i > 10000: break
    print(f"{num_bad_q} bad reads of {num_total} total")
    break
```
